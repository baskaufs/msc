```python
import os
import csv

def read_dict(filename):
    with open(filename, 'r', newline='', encoding='utf-8') as file_object:
        dict_object = csv.DictReader(file_object)
        array = []
        for row in dict_object:
            array.append(row)
    return array



def generate_query(alternatives):
    query = '''
select distinct ?item ?label where {
  VALUES ?value
  {
  ''' + alternatives + '''}
?item wdt:P1163 ?value.
?item rdfs:label ?label.
FILTER(lang(?label)='en')
  }
'''
    print(query)
```

```python
print(os.getcwd())
os.chdir('/Users/baskausj/github/rs.tdwg.org/format')
print(os.getcwd())

path = 'format.csv'
terms = read_dict(path)
print(terms)
```

```python

alternatives = ''
for term in terms:
    if '/' in term['controlled_value_string']:
        alternatives += '"' + term['controlled_value_string'] + '"\n'
alternatives = alternatives[:len(alternatives)-1]

query = generate_query(alternatives)
print(query)
```
