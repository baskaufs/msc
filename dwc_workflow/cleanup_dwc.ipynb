```python
import re
import requests   # best library to manage HTTP transactions
import csv        # library to read/write/parse CSV files
import json       # library to convert JSON to Python data structures
import pandas as pd

# -----------------
# Configuration section
# -----------------

# This is the base URL for raw files from the branch of the repo that has been pushed to GitHub
githubBaseUri = 'https://raw.githubusercontent.com/tdwg/rs.tdwg.org/dwc/'

# This is a Python list of the database names of the term version lists to be included in the document.
termLists = ['terms', 'iri', 'dc-for-dwc', 'dcterms-for-dwc', 'curatorial', 'dwcore', 'dwctype', 'geospatial', 'dwc-obsolete']
```

```python
# Create list of lists metadata table
for term_list_index in range(len(termLists)):
    # retrieve versions metadata for term list
    versions_url = githubBaseUri + termLists[term_list_index] + '-versions/' + termLists[term_list_index] + '-versions.csv'
    print(versions_url)
    versions_df = pd.read_csv(versions_url, na_filter=False)
    if term_list_index == 0:
        accumulated_frame = versions_df.copy()
    else:
        #print(accumulated_frame)
        #print(versions_df.copy())
        accumulated_frame = accumulated_frame.append(versions_df.copy(), sort=True)
accumulated_frame.reset_index(drop=True, inplace=True)
accumulated_frame.head()
```

```python
normative_frame = pd.read_csv('normative-document.csv', na_filter=False)
normative_frame.head()
```

```python
matched = []
norm_not_acc = []
for norm_row_index,norm_row in normative_frame.iterrows():
    print(norm_row['iri'])
    found = False
    for acc_row_index,acc_row in accumulated_frame.iterrows():
        if acc_row['version'] == norm_row['iri']:
            found = True
            matched.append(norm_row['iri'])
            break
    if not found:
        norm_not_acc.append(norm_row['iri'])
print('done')
```

```python
acc_not_norm = []
for acc_row_index,acc_row in accumulated_frame.iterrows():
    found = False
    for iri in matched:
        if acc_row['version'] == iri:
            found = True
            break
    if not found:
        print(acc_row['version'])
        acc_not_norm.append(acc_row['version'])
print('done')
```
